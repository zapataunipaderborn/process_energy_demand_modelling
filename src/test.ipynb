```python
# -----------------------------
# 1️⃣ Create richer synthetic log
# -----------------------------
random.seed(42)

cases = []

for case_id in range(1, 20):
    timestamp = datetime(2024, 1, 1, 8, 0, 0)

    case = f"Case_{case_id}"

    # Start
    cases.append({
        "case:concept:name": case,
        "concept:name": "A",
        "time:timestamp": timestamp
    })
    timestamp += timedelta(minutes=5)

    # Parallel behavior (B and C in random order)
    parallel_acts = ["B", "C"]
    random.shuffle(parallel_acts)

    for act in parallel_acts:
        cases.append({
            "case:concept:name": case,
            "concept:name": act,
            "time:timestamp": timestamp
        })
        timestamp += timedelta(minutes=5)

    # Choice between D and E
    choice = random.choice(["D", "E"])
    cases.append({
        "case:concept:name": case,
        "concept:name": choice,
        "time:timestamp": timestamp
    })
    timestamp += timedelta(minutes=5)

    # Optional loop on F
    if random.random() > 0.5:
        cases.append({
            "case:concept:name": case,
            "concept:name": "F",
            "time:timestamp": timestamp
        })
        timestamp += timedelta(minutes=5)

    # End
    cases.append({
        "case:concept:name": case,
        "concept:name": "G",
        "time:timestamp": timestamp
    })

df = pd.DataFrame(cases)
df = dataframe_utils.convert_timestamp_columns_in_df(df)
log = log_converter.apply(df)
```
